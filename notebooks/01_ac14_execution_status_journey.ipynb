## Governance Contract

- Purpose: expose the active repo control surface in one readable place
- Mode: governance-only, not proof
- Inputs: active docs and plan index
- Output: a compact summary of the current lane, follow-on lane, and latest empirical verdict


```python
from __future__ import annotations

from pathlib import Path
import json

def find_repo_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'pyproject.toml').exists() and (candidate / 'docs' / 'plans' / 'CLAUDE.md').exists():
            return candidate
    raise RuntimeError('Could not locate the AC14 repo root from the current working directory.')

REPO_ROOT = find_repo_root(Path.cwd())
TODO_PATH = REPO_ROOT / 'docs' / 'TODO.md'
NEXT24_PATH = REPO_ROOT / 'docs' / 'AC14_NEXT_24_HOURS.md'
PLAN_INDEX_PATH = REPO_ROOT / 'docs' / 'plans' / 'CLAUDE.md'
DECISION_PATH = REPO_ROOT / '.ac14_out' / 'full_trials_gate_1' / 'experiment_decision.json'
assert TODO_PATH.exists()
assert NEXT24_PATH.exists()
assert PLAN_INDEX_PATH.exists()
assert DECISION_PATH.exists()
decision_artifact = json.loads(DECISION_PATH.read_text())
decision_artifact['verdict']
```

## Status Readout

- Purpose: prove the active docs exist and agree on the current lane
- Input -> output: TODO/NEXT24/plan index -> one compact governance summary
- Acceptance criteria: the docs point to Plan #63 as the active empirical contract, keep Plan #61 completed, and do not drift back to stale smoke-repair lanes
- Status: `proven`
- Execution mode: `live`


```python
todo_text = TODO_PATH.read_text()
next24_text = NEXT24_PATH.read_text()
plan_index_text = PLAN_INDEX_PATH.read_text()

assert 'Plan #63: Runtime-First Comparison Contract' in todo_text
assert 'Plan #61: Executable Journey Notebook Remediation' in todo_text
assert 'Plan #63: Runtime-First Comparison Contract' in next24_text
assert 'Plan #61: Executable Journey Notebook Remediation' in next24_text
assert '| 61 | [Executable Journey Notebook Remediation]' in plan_index_text
assert '| 63 | [Runtime-First Comparison Contract]' in plan_index_text

governance_summary = {
    'active_plan': '63_runtime_first_comparison_contract',
    'completed_notebook_plan': '61_executable_journey_notebook_remediation',
    'latest_empirical_verdict': decision_artifact['verdict'],
}
governance_summary
```
